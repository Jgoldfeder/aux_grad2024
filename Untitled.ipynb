```python
import matplotlib.pyplot as plt

x = [0, 0.05, 0.1, 0.15, 0.2, 0.25,0.3]

y1 = [93.9,55.8,58.1,56.2,53.9,50.5,45.6]
y2 = [93.8,92.1,89.8,81.1,73.2,65.6,56]

plt.plot(x, y1)
plt.plot(x, y2, '-.')

plt.xlabel("eps")
plt.ylabel("acc")
plt.title('attack')
plt.show()
```
